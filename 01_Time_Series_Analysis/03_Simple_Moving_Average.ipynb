```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("applestock_data.csv")
df.head()
```

```python
df[" High"][0]
```

```python
df["Date"]=pd.to_datetime(df["Date"])
```

```python
df.set_index("Date",inplace=True)
```

```python
df.columns
```

```python
for i in [' Close/Last', ' Open', ' High', ' Low']:
    df[i]=df[i].apply(lambda x : float(x.replace(" $","")))
```

```python
pd.set_option("display.max_rows",None)
```

```python
df["rolling:5days_high"]=df[" High"].rolling(window=5).mean()
```

```python
df["rolling:10days_high"]=df[" High"].rolling(window=10).mean()
```

```python
df["rolling:15days_high"]=df[" High"].rolling(window=15).mean()
```

```python
df["rolling:30days_high"]=df[" High"].rolling(window=30).mean()
```

```python

df[[" High","rolling:5days_high","rolling:10days_high","rolling:15days_high","rolling:30days_high"]].plot(figsize=(14,6))
plt.show()
```
