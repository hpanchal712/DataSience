```python
# Creating a pdf file object
pdf = open('test.pdf','rb')

# Create pdf reader object
pdf_reader = PyPDF2.PdfReader(pdf)

# checking number of pages in a pdf file
print('Number of pages: ', len(pdf_reader.pages))

# Creating a page object
page = pdf_reader.pages[4]

# Finally extracting text from the page
print(page.extract_text())

# closing the pdf file
pdf.close()
```

```python
# Reading the PDF from an URL
import urllib.request
from io import BytesIO
wFile = urllib.request.urlopen('http://www.udri.org/pdf/02%20working%20paper%201.pdf')
pdfreader = PyPDF2.PdfReader(BytesIO(wFile.read()))
```

```python
len(pdfreader.pages)
```

```python
# Extracting page#3 of the document
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

pageObj = pdfreader.pages[2]
page3 = pageObj.extract_text()

page3
```

```python
# Cleaning the text

# Tokenize the words
tokens = word_tokenize(page3)
tokens
```

```python
# Now let's remove stopwords and punctuations
stop_words = stopwords.words('english')
punctuations = ['(',')',';',':','[',']',',','...','.']

keywords = [word for word in tokens if not word in stop_words and not word in punctuations]
```

```python
keywords
```

```python
len(tokens)
```

```python
name_list=list()

check = ['Mr.', 'Mrs.', 'Ms.']

for idx, token in enumerate(tokens):
    #print(idx, token)  --> 80 Mr.   81 Jagdish    82 Talreja
    if token.startswith(tuple(check)) and idx < (len(tokens)-1): # idx < 276-1 => idx < 275 (275 is the last index)
        name = token + tokens[idx+1] + ' ' + tokens[idx+2]
        name_list.append(name)
        
print(name_list)
```
